```python
most_goals_index = players['Goals'].idxmax()
player_most_goals = players.loc[most_goals_index]

print("Player with the most goals:")
print(player_most_goals)
```

```python
least_goals_index = players['Goals'].idxmin()
player_least_goals = players.loc[least_goals_index]
print("Player with the least goals:")
print(player_least_goals)
```

```python
most_caps_index = players['Caps'].idxmax()
player_most_caps = players.loc[most_caps_index,['Name','Caps']]

print("Player with the most caps:")
print(player_most_caps)
```

```python
most_Club_index = players['Club'].idxmax()
player_most_Club = players.loc[most_caps_index]

print("Player with the most Club:")
print(player_most_Club)
```

```python
foot = players['Foot'].value_counts()
plt.figure(figsize=(10, 6))
foot.plot(kind='bar')
plt.title('Distribution of Player according to their dominant foot')
plt.xlabel('Foot')
plt.ylabel('Number of Players')
plt.show()
```

```python
players_grouped = players.groupby('Country').sum().reset_index()
plt.figure(figsize=(15, 10))  
goalplot = sns.barplot(x='Country', y='Goals', data=players_grouped)
plt.title('Total Number of Goals Scored by Players of a Country')
plt.xlabel('Name of Country')
plt.ylabel('Number of Goals')

for p in goalplot.patches:
    goalplot.annotate(format(p.get_height(), '.1f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 10), 
                      textcoords='offset points')

plt.xticks(rotation=90)  
plt.show()
```

```python
data_group = players.groupby('Club').sum().reset_index()
top_10_clubs = data_group.sort_values(by='MarketValue', ascending=False).head(10)
plt.figure(figsize=(10, 15))
sns.barplot(x='MarketValue',y='Club',data = top_10_clubs)
plt.title('Total Market Value of Players in a club (Top 10)')
plt.xlabel('Total Market Value')
plt.ylabel('Name of Clubs')
plt.show()
```
